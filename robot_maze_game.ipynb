```python
import random

grid_size = 5


def generate_positions():
    start = (0, 0)

    positions = []
    for y in range(grid_size):
        for x in range(grid_size):
            if (x, y) != start:
                positions.append((x, y))

    goal = random.choice(positions)
    positions.remove(goal)
    traps = random.sample(positions, 4)

    return goal, traps


def display_grid(robot_x, robot_y, goal, traps):
    for y in range(grid_size):
        row = []
        for x in range(grid_size):
            if (x, y) == (robot_x, robot_y):
                row.append("R")
            elif (x, y) == goal:
                row.append("G")
            elif (x, y) in traps:
                row.append("X")
            else:
                row.append(".")
        print(" ".join(row))


def play_game():
    robot_x = 0
    robot_y = 0
    steps = 0

    goal, traps = generate_positions()

    while True:
        print("\nRobot Navigation Simulator")
        print(f"Steps: {steps}")
        display_grid(robot_x, robot_y, goal, traps)

        move = input("\nEnter move (up, down, left, right): ").strip().lower()

        if move not in ["up", "down", "left", "right"]:
            print("Invalid command. Please use up, down, left, or right.")
            continue

        new_x = robot_x
        new_y = robot_y

        if move == "up":
            new_y -= 1
        elif move == "down":
            new_y += 1
        elif move == "left":
            new_x -= 1
        elif move == "right":
            new_x += 1

        if new_x < 0 or new_x >= grid_size or new_y < 0 or new_y >= grid_size:
            print("You cannot move outside the grid.")
            continue

        robot_x = new_x
        robot_y = new_y
        steps += 1

        if (robot_x, robot_y) in traps:
            print("\nGame over! You stepped on a trap.")
            print(f"Total steps: {steps}")
            display_grid(robot_x, robot_y, goal, traps)
            break

        if (robot_x, robot_y) == goal:
            print("\nCongratulations! You reached the goal.")
            print(f"Total steps: {steps}")
            display_grid(robot_x, robot_y, goal, traps)
            break


while True:
    play_game()

    play_again = input("\nDo you want to play again? (yes/no): ").strip().lower()

    if play_again != "yes":
        print("Thanks for playing!")
        break
```
